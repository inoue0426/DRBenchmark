```python
def get_mask(PATH):
    name = pd.read_csv("data/masked_cancer_gene/Cline_to_TumourType.csv")
    tmp = pd.read_csv(f"../{PATH}/cell line_GEP.csv")
    cell_lines = sorted(set(tmp["SAMPLE_ID"]) & set(name["CELL_LINE_NAME"]))
    name = name[name.CELL_LINE_NAME.isin(cell_lines)]
    mask = pd.read_csv("data/masked_cancer_gene/masked.csv")
    mask = mask[mask.SAMPLE_ID.isin(name.COSMIC_ID)].loc[:, tmp.columns]
    conv = dict(name[["COSMIC_ID", "CELL_LINE_NAME"]].values)
    mask["SAMPLE_ID"] = [conv[i] for i in mask["SAMPLE_ID"]]
    missing_cells = list(set(tmp.SAMPLE_ID) - set(mask.SAMPLE_ID))
    df = pd.DataFrame(
        np.ones((len(missing_cells), mask.shape[1])), columns=mask.columns
    )
    df["SAMPLE_ID"] = missing_cells
    pd.concat([mask, df]).sort_values("SAMPLE_ID").reset_index(drop=True).to_csv(
        f"../{PATH}/masked.csv", index=False
    )
    tmp.sort_values("SAMPLE_ID").reset_index(drop=True).to_csv(
        f"../{PATH}/cell line_GEP.csv", index=False
    )
```

```python
for i in ["nci_data", "gdsc1_data", "gdsc2_data", "ctrp_data"]:
    get_mask(i)
```
